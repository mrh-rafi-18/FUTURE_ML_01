```python

import pandas as pd
from prophet import Prophet
import holidays
import matplotlib.pyplot as plt



monthly_sales = df.groupby(pd.Grouper(key='Order Date', freq='ME'))['Sales'].sum().reset_index()
monthly_sales.rename(columns={'Order Date':'ds', 'Sales':'y'}, inplace=True)

print("Monthly aggregated sales:")
print(monthly_sales.head())


```

```python

model = Prophet()
model.add_country_holidays(country_name='US')


model.fit(monthly_sales)

future = model.make_future_dataframe(periods=12, freq='ME')
forecast = model.predict(future)


fig1 = model.plot(forecast)
plt.title("Monthly Sales Forecast")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()


fig2 = model.plot_components(forecast)
plt.show()

```

```python
print("Next 12 months forecast:")
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(12))

# 12️⃣ Export forecast to Kaggle working directory
export_df = forecast[['ds', 'yhat', 'yhat_upper', 'yhat_lower']].rename(columns={
    'ds': 'Month',
    'yhat': 'Forecasted Sales',
    'yhat_upper': 'Forecasted Sales Upper',
    'yhat_lower': 'Forecasted Sales Lower'
})
export_df.to_csv('/kaggle/working/superstore_forecast.csv', index=False)

print("Forecast exported to '/kaggle/working/superstore_forecast.csv'")
```
